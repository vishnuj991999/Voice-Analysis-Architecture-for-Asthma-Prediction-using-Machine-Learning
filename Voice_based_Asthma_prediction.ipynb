```python
df=pd.read_csv('/content/asthma_RBL.csv')
df.head()
```

```python
df['TARGET VALUE'].value_counts()
```

```python
X = df.drop(columns=['TARGET VALUE'])
y = df['TARGET VALUE']

# Create SMOTE object for balancing the classes
smote = SMOTE(sampling_strategy={0: 200, 1: 200}, random_state=42)

# Apply SMOTE
X_resampled, y_resampled = smote.fit_resample(X, y)

# Create a DataFrame for the resampled data
df2 = pd.DataFrame(X_resampled, columns=X.columns)
df2['TARGET VALUE'] = y_resampled

# Print the resampled DataFrame shape
print("Original data shape:", df.shape)
print("Resampled data shape:", df2.shape)
```

```python
df2['TARGET VALUE'].value_counts()
```

```python
df2.replace([np.inf, -np.inf], np.nan, inplace=True)
df2.fillna(999, inplace=True)
```

```python
features=df2.loc[:,df.columns!='TARGET VALUE'].values[:,1:]

labels=df2.loc[:,'TARGET VALUE'].values

print(labels[labels==1].shape[0], labels[labels==0].shape[0])


scaler=MinMaxScaler((-1,1))
```

```python
x=scaler.fit_transform(features)

y=labels

x_train,x_test,y_train,y_test=train_test_split(x, y, test_size=0.2, random_state=7)
```

```python
#logisticregression

model1 = LogisticRegression()
model1.fit(x_train,y_train)
y_predict_LR = model1.predict(x_test)

from sklearn.metrics import confusion_matrix

pd.DataFrame(

    confusion_matrix(y_test,y_predict_LR),

    columns=['ASTHMA NEGATIVE', 'PREDICTED ASTHMA'],

    index=['True Healthy', 'True Asthma']

)

```

```python
print(metrics.classification_report(y_test, y_predict_LR, labels=[1, 0]))
```

```python
#KNN
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 10) # n_neighbors = k
knn.fit(x_train,y_train)

y_predict_knn= knn.predict(x_test)

from sklearn.metrics import confusion_matrix

pd.DataFrame(

    confusion_matrix(y_test,y_predict_knn),

    columns=['ASTHMA NEGATIVE', 'PREDICTED ASTHMA'],

    index=['True Healthy', 'True Asthma']

)
```

```python
print(metrics.classification_report(y_test, y_predict_knn, labels=[1, 0]))
```

```python
#Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 6)
regressor.fit(x_train, y_train)
y_pred_DR = regressor.predict(x_test)
pd.DataFrame(

    confusion_matrix(y_test,y_pred_DR ),

    columns=['ASTHMA NEGATIVE', 'PREDICTED ASTHMA'],

    index=['True Healthy', 'True Asthma']

)
```

```python
print(metrics.classification_report(y_test,y_pred_DR, labels=[1, 0]))
```
